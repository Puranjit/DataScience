# Data Visualization (Part - C)

`Seaborn` is a data visualization library built on top of matplotlib and closely integrated with pandas data structures in Python.<br> 
`Seaborn` helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.<br>

Using `Seaborn` we can plot wide varieties of plots like:
- Distribution plots
- Pie chart and Bar chart
- Scatter plots
- Pair plots
- Heat maps

Seaborn is extensively used by Data scientists to perform certain data analysis by plotting different plots to extract useful information such as relationships, trends, finding outliers and co-relations between different variables in various data science projects.

To learn more about Seaborn implementation and usage refer to this [Link](https://seaborn.pydata.org/introduction.html)<br>


```python
# Packages/libraries that need to be installed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Loading the raw_data

```python
raw_data = pd.read_csv('Marketing Raw Data.csv')
print(raw_data.shape)
raw_data.head()
```

```python
raw_data.tail()
```

### 2. Line graph

```python
# Investigating revenue by date

ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', data = raw_data)
```

```python
raw_data.loc[0:15]
```

```python
# Adding categories in the line graph
ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', hue = 'Promo', data = raw_data)
```

```python
# Resizing plot
sns.set(rc = {'figure.figsize' : (12,10)})

# Adding categories in the line graph
ax = sns.lineplot(x = 'Week_ID', y = 'Revenue', hue = 'Promo', data = raw_data, style = 'Promo')
```

## 3. Bar plots

```python
# Revenue by month
ax = sns.barplot(x = 'Month_ID', y = 'Revenue', data = raw_data)

raw_data[['Month_ID', 'Revenue']].groupby('Month_ID', as_index = False).agg({'Revenue':'mean'})
```

```python
# Total revenue by month - removing CI
ax = sns.barplot(x = 'Month_ID', y = 'Revenue', ci = False, data = raw_data)
```

```python
# Revenue/month and including promo as a factor too
ax = sns.barplot(x = 'Month_ID', y = 'Revenue', ci = False, data = raw_data, hue = 'Promo')
```

```python
ax = sns.barplot(x = 'Revenue', y = 'Year', data = raw_data, ci = False, hue = 'Promo', orient = 'h', color = '#71714B')
```

## 4. Histograms

```python
# Storing all the revenue-values in an array
x = raw_data['Revenue'].values
x
```

```python
ax = sns.distplot(x, color = 'green')
```

```python
ax = sns.displot(x, color  ='blue')
```

```python
ax = sns.histplot(x, color  ='orange')
```

```python
# Calculating mean and showing it in the histogram
mean = raw_data['Revenue'].mean()
print(mean)
ax = sns.histplot(x, color = 'green')

plt.axvline(mean, 0,1, color = 'red')
plt.show()
```

## 5. Box plots

```python
x = raw_data['Revenue'].values

ax = sns.boxplot(x)
print('The median is: ', raw_data['Revenue'].median())
```

```python
ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = raw_data)
```

```python
ax = sns.boxplot(x = 'Day_Name', y = 'Revenue', data = raw_data, color  ='#E3E349')
ax = sns.swarmplot(x = 'Day_Name', y = 'Revenue', data = raw_data, color = 'red')
```

## 6. SubPlots

```python
# Setting up matplotlib figure
fig, axes = plt.subplots(2,2, figsize = (12,7))

a = raw_data['Revenue'].values
b = raw_data['Visitors'].values
c = raw_data['Marketing Spend'].values

# plot 1
sns.histplot(a, color = 'blue', ax = axes[0,0])

# plot 2
sns.histplot(b, color = 'blue', ax = axes[0,1])

# plot 3
sns.histplot(c, color = 'blue', ax = axes[1,0])

# plot 4
sns.boxplot(x = 'Revenue', y = 'Day_Name', color = '#ABAB16', data = raw_data, ax = axes[1,1])
sns.swarmplot(x = 'Revenue', y = 'Day_Name', color = 'red', data = raw_data, ax = axes[1,1])
```

# 8. Heat Map

### Co-relation
.corr() is used to find the pairwise correlation of all columns in the dataframe. Any null values are automatically excluded
The closer to 1 or -1 the better. As one variable increases, the other variable tends to also increase / decrease
More Info here: [Reference](https://statisticsbyjim.com/basics/correlations/)

```python
raw_data.corr()
```

```python
ax = sns.heatmap(raw_data.corr())
```

```python
pc = raw_data[['Revenue','Visitors','Marketing Spend', 'Promo']].corr(method ='pearson')

cols = ['Revenue CC','Visitors CC','Marketing Spend CC']

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 35},
                 cmap = 'BuPu')
```

To learn more about Heat maps refer to this [Reference](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
