# Anaconda Tutorial
Anaconda is a free Python distribution specifically designed for scientific computing (data science, machine learning). Anaconda is popular because it brings many of the tools used in data science and machine learning with just one install, so it's great for having short and simple setup. Anaconda uses the concept of creating environments so as to isolate different libraries and versions.

### Use of multiple environments
- Manages the correct versions and dependenceies for your project
- Separates different installations versions in different environments

## Install
Go to (https://www.anaconda.com/products/individual-d) and download the latest installer for your machine using the instructions

## Get started

### General commands

Check version<br>
`conda --version`<br>
Check more information<br>
`conda info`<br>
Update conda<br>
`conda update conda`<br>

### Manage environments

Create a virtual environment<br>
`conda create -name myenv`<br>
Specify a specific Python version  
`conda create -name myenv Python=3.8`   
Specify specific packages that are installed  
`conda create -name myenv Python=3.8 numpy matplotlib`  

Activate it (Depending on your machine):  
`conda activate myenv`  

Your terminal will feature the current activated environment.  

Deactivate it (Depending on your machine):  
`conda deactivate`  

List all environments:  
`conda env list`

Remove a specific environment  
`conda env remove -name myenv`

### Manage packages
Install specific packages  
`conda install numpy`  
Install multiple packages with one command  
`conda install seaborn matplotlib pandas`

Deinstall packages  
`conda remove numpy`  

List all packages in an environment  
`conda list`  

Update a package  
`conda update numpy`

Search for packages  
`conda search numpy`

Installing with pip is also possible:  
`pip install numpy`  
It is recommended to install pip in your environment and then use your local pip version. Otherwise it will try to fall back to other ones on the machine  
`conda install pip`  
`pip install numpy`
