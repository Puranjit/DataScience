```python
# Assume : Irrigation on Day 1 = 0
Irr = 0

# Storing Values whether irrigation was done or not?
IrB = []
IrSumB = [Irr]

# Last date of irrigation - Apply irrigation if LD <=3
LD = []

# Allowable depletion
AD = 3.5

# Initial Soil water depletion = 0
SWD = [0]

# Running for loop
for i in range(len(df)):
    
    SWD.append(SWD[i] + df.ETSum[i] - df.PrepSum[i] - IrSumB[i])
    
    LD.append((AD - SWD[i])/ETfor[i])
    
    if LD[i] <= 3:
        IrB.append(1)
        
        Irr = Irr + 1
        
    else:
        IrB.append(0)
    
    IrSumB.append(Irr)
```

```python
PISumB = []
for i in range(len(df)):
    PISumB.append(IrSumB[i] + df.PrepSum[i])
PISumB
```
