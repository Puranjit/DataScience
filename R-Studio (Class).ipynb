```python
# Importing libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the csv file
df = pd.read_csv('Havelock_part1_dat_UTM.csv')
df
```

We need to remove following columns as they are not extensively used for the following parts
* smp_id
* Lab No
* Date Recd 
* Date Rept
* B Depth
* E Depth
* WDRF Buffer pH
* 1:1 S Salts mmho/cm
* Excess Lime
* Texture No
* lbs N/A
* Potassium ppm K
* Magnesium ppm Mg 
* Sodium ppm Na
* %H Sat
* %K Sat
* %Ca Sat
* %Mg Sat
* Na Sat

```python
# Line 1-4 in R-code
df = df.drop(['smp_id', 'Lab No', 'Date Recd', 'Date Rept', 'B Depth', 'E Depth', 'WDRF Buffer pH', '1:1 S Salts mmho/cm', 
              'Excess Lime', 'Texture No', 'lbs N/A', 'Potassium ppm K', 'Magnesium ppm Mg', 'Sodium ppm Na', '%H Sat', 
              '%K Sat', '%Ca Sat', '%Mg Sat', '%Na Sat'], axis = 1)

df.head()
```

```python
# Heatmap in Seaborn library gives us an idea on how different variables that are present in our data are linked to each other
plt.figure(figsize=(10,8))
ax = sns.heatmap(df.corr(), vmax = 1, vmin = -1, annot= True)
```

```python
# Line 5 in R-code

# Finding 
ax = sns.pairplot(df, plot_kws={'color':'green'})
```

```python
df
```

```python
# Line 7-8 in R-code

# Renaming columns in the dataframe
df = df.rename(columns={'UTM Northing':'Lat', 'UTM Easting':'Lon', 'Elevation':'Ele', '1:1 Soil pH':'ph', 'Nitrate-N ppm N':'No3', 
                        'Calcium ppm Ca':'Ca', 'CEC/Sum of Cations me/100g':'CEC', 'Mehlich P-III ppm P':'P', 'Total N ppm':'TN', 
                        'Soil Moisture %':'MC', 'Total Carbon %':'TC'})
df
```

```python
# Line 9 in R-code

# Plotting barplot of pH values in the Dataframe
plt.figure(figsize=(8,6))

plt.hist(df.ph)

plt.title('pH values')
plt.xlabel('pH Value')
plt.ylabel('Frequency')

plt.show()
```

```python
# Storing all pH values in a variable from DataFrame
pH = df.ph
```

```python
# Line 10 in R-code

## Finding mean of all pH values in the Dataframe
mean_pH = np.mean(pH)
mean_pH
```

```python
mean_pH = round(mean_pH, 3)
mean_pH
```

```python
# Line 11 in R-code

# Finding max pH value in column
max_pH = np.max(pH)
max_pH
```

```python
# Line 12 in R-code

# Finding min pH value in column
min_pH = np.min(pH)
min_pH
```
