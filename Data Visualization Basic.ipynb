```python
plt.figure(figsize = (10,8))

plt.plot(data.NationalMarketingYear, data.Total_Supply)

plt.xticks([x for x in range(int(data.NationalMarketingYear[0]), int(data.NationalMarketingYear[0]) + len(data.NationalMarketingYear), 3)])

# Plotting X & Y-axis
plt.xlabel('Years')
plt.ylabel('Total Supply (Million Tonnes)')
plt.title('Total Supply of maize in the US')

plt.show()
```

```python
plt.figure(figsize = (10,8))

plt.plot(data.NationalMarketingYear, data.Population)

plt.xticks([x for x in range(int(data.NationalMarketingYear[0]), int(data.NationalMarketingYear[0]) + len(data.NationalMarketingYear), 2)])

# Plotting X & Y-axis
plt.xlabel('Years')
plt.ylabel('Population (Million)')
plt.title('Population of the US over the years')

plt.show()
```

```python
data[['Total_Utilization', 'Domestic_Utilization']]
```

```python
plt.figure(figsize = (10,12))

plt.plot(data.NationalMarketingYear, data.Total_Utilization)
plt.plot(data.NationalMarketingYear, data.Domestic_Utilization)

plt.xticks([2000, 2003, 2006, 2009, 2012, 2015, 2018, 2021])
# plt.xticks([x for x in range(int(df.NationalMarketingYear[0]), int(df.NationalMarketingYear[0]) + len(df.NationalMarketingYear), 2)])

# Plotting X & Y-axis
plt.xlabel('Years')
plt.ylabel('Utilization (Million Tonnes)')
plt.title('Total Utilization Vs Domestic utilization of maize in the US', fontdict={'fontsize':18})

# Saving plot 
plt.savefig('Totalutilization Vs DomesticUtilization.png', dpi = 300)
# plt.legend()

plt.show()
```
