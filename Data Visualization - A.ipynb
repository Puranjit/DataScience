### matplotlib.pyplot

`matplotlib.pyplot` is a state-based interface to `matplotlib`. It provides a MATLAB-like way of plotting. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.<br>
pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation.<br>

### Use of Legend in a plot

```python
x = [i for i in range(5)]
y = [i**2 for i in range(5)]

print(x)
print(y)

# plt.plot(x,y, label = 'y = x ** 2')
# plt.plot(x,y, label = 'y = x ** 2', c = 'red')
# plt.plot(x,y, linewidth = 3.5, c = 'r', label = 'y = x**2', linestyle = 'dashed')
plt.plot(x,y, linewidth = 3.5, c = 'r', label = 'y = x**2', linestyle = 'dashed', marker = 'o', markersize = 12, markeredgecolor = 'b')

# Labelling X & Y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding a legend
plt.legend()

plt.show()
```

You can check the different `markers` and `colors` that are available for use in these use-case scenarios [Link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) towards the bottom.

```python
x = [i for i in range(5)]
y = [i**2 for i in range(5)]

plt.figure(figsize=(8,8))

print('x is: ', x)
print('y is: ', y)

y1 = [i*2 for i in range(5)]
print('y1 is: ', y1)

# plt.plot(x,y, label = 'y = x ** 2')
# plt.plot(x,y1, label = 'y = x ** 2')

plt.plot(x,y, label = 'y = x**2', linewidth = 2, c = 'r', linestyle = '--', marker = 'o', markersize = 8, markeredgecolor = 'b')
plt.plot(x,y1, label = 'y1 = x*2', linewidth = 2, marker = '^', markersize = 8, markeredgecolor = 'g')

# Labelling X & Y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Selecting intervals that we want to include on our axis
plt.xticks([0,1,2,3,4])
plt.yticks([0,4,8,12,16])

# Adding a title specifying its location above graph
plt.title('Graph on List comprehensions', loc = 'center')

# Saving the plot as a figure
plt.savefig('mygraph2.png', dpi = 200)

# Adding a legend
plt.legend()
plt.show()
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 2, 3]

# Plotting a bar chart
plt.bar(labels, values)
```

```python
labels = ['A', 'B', 'C', 'D']
values = [1, 2, 3, 4]

plt.figure(figsize = (6,5))

# Plotting a bar chart
bars = plt.bar(labels, values)

# Showing special patterns for different labels
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('-')
bars[3].set_hatch('+')
```

```python
labels = ['A', 'B', 'C', 'D']
values = [1, 2, 3, 4]

plt.figure(figsize = (6,5))

# Plotting a bar chart
bars = plt.bar(labels, values)

# Showing special patterns for different labels
patterns = ['/', '*', '-', '+']
for i in range(len(patterns)):
    bars[i].set_hatch(patterns.pop())
#     bars[i].set_hatch(patterns.pop(0))
#     bars[i].set_hatch(patterns[i])
```
