```python
# Taking sum of consecutive values and forming a new column

# Defining an empty list to store all the consecutive sum values in the new list created 
z = []

# Initialize sum1 (varibale to 0)
sum1 = 0

# Storing all values from `York County` in the variable x
x = df['York County']

# Creating a loop for this purpose
for i in x:
    sum1 = sum1 + i
    z.append(sum1)
```

```python
z
```

```python
df['YorkPrecipSum'] = z
df.tail()
```

### Plotting graphs using `matplotlib`

```python
df = df.rename(columns={'County' : 'Month'})
df.head()
```

```python
plt.figure(figsize = (10,8))

plt.plot(df.Month, df['York County'], label = 'Precip in York')
plt.plot(df.Month, df['YorkPrecipSum'], label = 'Sum(Precip) in York')

plt.xlabel('Months')
plt.ylabel('Precipitation (inches)')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,8))

plt.plot(df.Month, df['YorkPrecipSum'], 'r-',linewidth = 5, label = 'Sum(Precip) in York')
plt.bar(df.Month, df['YorkPrecipSum'], label = 'Sum(Precip) in York', align = 'center')
plt.xlabel('Months')
plt.ylabel('Precipitation (inches)')

plt.legend()
plt.show()
```
